Agora vamos organizar o novo csv para deixarmos compatível com o "oscarwinners" e uni-los.

```python
import pandas as pd
import numpy as np
```

```python
df_novo = pd.read_csv('datasets/oscar_vencedores_2014_2024.csv')
```

```python
mapa_categorias = {
    'ACTOR IN A LEADING ROLE': 'Best Actor',
    'ACTOR IN A SUPPORTING ROLE': 'Best Supporting Actor',
    'ACTRESS IN A LEADING ROLE': 'Best Actress',
    'ACTRESS IN A SUPPORTING ROLE': 'Best Supporting Actress',
    'DIRECTING': 'Best Director'
}

df_novo['category'] = df_novo['category'].replace(mapa_categorias)
print(df_novo['category'].head())
```

```python
df_novo = df_novo.rename(columns={
    'film': 'movie',
    'year_ceremony': 'year_edition'
})
```

```python
colunas = [
    'name', 
    'birth_year', 
    'birth_date', 
    'birthplace', 
    'race_ethnicity', 
    'religion', 
    'sexual_orientation', 
    'year_edition', 
    'category', 
    'movie' ]

df_novo_padronizado = df_novo.reindex(columns=colunas)
```
